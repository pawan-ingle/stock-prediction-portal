```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Type of Plots

### Line Chart

```python
x = [1,2,3,4,5]
y = [10,20,25,30,40]
```

```python
plt.plot(x,y);
plt.title("Simple Line Plot");
plt.xlabel('X-axis label');
plt.ylabel('Y-axis label');
plt.grid(True)
```

### Bar Chart

```python
months = ['Jan', 'Feb', 'Mar', 'Apr']
revenue = [100, 200, 300, 500]
```

```python
plt.figure(figsize = (10,6))
plt.bar(months, revenue);
plt.title("Monthly Revenue Bar Chart");
plt.xlabel('Months');
plt.ylabel('Revenue');
```

### Scatter Plot

```python
heights = [65, 70, 68, 72, 66]
weights = [150, 160, 155, 180, 145]
```

```python
plt.scatter(heights, weights)
plt.title('Heights Vs. Weights Scatter Plot')
plt.xlabel("Heights")
plt.ylabel("Weights")
```

### Subplots and Legends

```python
days = [1,2,3,4,5]
original_price = [10, 20, 25, 30, 40]
predicted_price = [8, 21, 25, 28, 38]
```

```python
plt.plot(days, original_price, 'b-o', label="original_price")
plt.plot(days, predicted_price, 'g-o', label="predicted_price")
plt.legend()
plt.title("Plot with Legends")
plt.xlabel("Weekdays")
plt.ylabel("Price")
```

### SubPlots

```python
# First Subplot - original price
plt.subplot(3,1,1)
plt.plot(days, original_price, 'b-o', label="original_price")
plt.title("Original price over time")
plt.xlabel("Weekdays")
plt.ylabel("Price")
plt.legend()
```

```python
# Second Subplot - predicted price
plt.subplot(3,1,2)
plt.plot(days, predicted_price, 'g-o', label="predicted_price")
plt.title("Predicted price over time")
plt.xlabel("Weekdays")
plt.ylabel("Price")
plt.legend()
```

```python
# Third Subplot - Diff between original & predicted price
price_difference = [o - p for o, p in zip(original_price, predicted_price)]
plt.subplot(3,1,3)
plt.bar(days, price_difference, color ='purple', label="price_difference")
plt.title("Diff between original & predicted price")
plt.xlabel("Days")
plt.ylabel("Price")
plt.legend()
```

### Saving the Figure

```python
plt.plot(days, original_price, 'b-o', label="original_price")
plt.plot(days, predicted_price, 'g-o', label="predicted_price")
plt.legend()
plt.title("Plot with Legends")
plt.xlabel("Weekdays")
plt.ylabel("Price")
plt.savefig("example_prediction.png", dpi=300) # save the image
```
